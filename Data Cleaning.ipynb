```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the dataset
df = pd.read_csv('advertising.csv - advertising.csv.csv')


# check for null values
print(df.isnull().sum())
```

```python
df # before Removing null values,no null values in data set
```

```python
# remove null values
df = df.dropna()
df
```

```python
df.boxplot(column=['TV', 'Radio', 'Newspaper'])
plt.show()

```

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1
upper_bound = q3 + 1.5 * iqr
lower_bound = q1 - 1.5 * iqr
```

```python
df = df[(df > lower_bound) & (df < upper_bound)]
df
```

```python
df.boxplot()
plt.show()
```

```python
df.to_csv('Cleaned_advertising.csv')
```
