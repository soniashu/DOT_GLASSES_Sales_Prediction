## Outlier Analysis

```python
fig, axs = plt.subplots(3, figsize = (5,5))
plt1 = sns.boxplot(df['TV'], ax = axs[0])
plt2 = sns.boxplot(df['Newspaper'], ax = axs[1])
plt3 = sns.boxplot(df['Radio'], ax = axs[2])
plt.tight_layout()
```

<b>There are no considerable outliers present in data

<h3><b>Sales(Target)</h3>

```python
sns.boxplot(df['Sales'])

plt.show()
```

<b>There is no outlier present in target Sales<br>
There is average sales of about 15,000 dollars

<h3>Features (Predictors)</h3>

```python
# to check all variables at once

df.hist( bins = 10, figsize = (10,10))
plt.show()
```

<h4>Spending budget for advertisments on TV is in all range, same for Radio but less than TV. But on newspapers, less budget is spend</h4>

<h4 style = "color : blue">Let's check with values</h4>

```python
# Calculate correlations
sns.pairplot(df)
```

<h4>
Target Sales is directly correlated with TV, little with Radio but there is no correlation with Newspaper

There is no correlation seems between all features.

Spending budget for advertisments on TV is in all range, same for Radio but less than TV. But on newspapers, less budget is spent</h4>

```python
# scatter plot

fig, axs = plt.subplots(1, 3)
df.plot(kind='scatter', x='TV', y='Sales', ax=axs[0], figsize=(16, 7), color = 'red')
df.plot(kind='scatter', x='Radio', y='Sales', ax=axs[1], color = 'orange')
df.plot(kind='scatter', x='Newspaper', y='Sales', ax=axs[2], color = 'purple')
```

```python
# Calculate correlations
sns.heatmap(df.corr(),annot=True)
```

<h4>As is visible from the pairplot and the heatmap, the variable TV seems to be most correlated with Sales</h4>
